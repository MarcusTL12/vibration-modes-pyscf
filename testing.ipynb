```julia
using PyCall
pyscf = pyimport("pyscf")
geomopt = pyimport("pyscf.geomopt.geometric_solver")
```

```julia
function searchkeys(s, o)
    filter(startswith(s), String.(keys(o)))
end
```

```julia
mol = pyscf.gto.M(atom="N 0 0 0; H 1 0 0; H 0 1 0; H 0 0 1"; basis="sto3g")
```

```julia
hf = mol.RHF().run()
```

```julia
mol_opt = geomopt.optimize(hf)
```

```julia
hf_opt = mol_opt.RHF().run()
```

```julia
h = hf_opt.Hessian().hess()
```

```julia
using LinearAlgebra
```

```julia
hp = PermutedDimsArray(h, (3, 1, 4, 2))
```

```julia
hm = reshape(hp, 12, 12)
```

```julia
e, v = eigen(hm)
```
